```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df=pd.read_csv('company_sales_data.csv')
df.head()
```

```python
plt.figure(figsize=(8,6))
month_number=df['month_number'].tolist()
total_profit=df['total_profit'].tolist()
plt.plot(month_number,total_profit)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Toot paste sales data')
plt.figure(figsize=(10,20))

plt.show()
```

```python
x=df['month_number'].tolist()
y=df['total_units'].tolist()
plt.plot(x,y,label='Profit Data Of Last Year',color='r',marker='o',mfc='k',linestyle='--',linewidth='2')
plt.xlabel('Month Number')
plt.ylabel('Total units')
plt.title('Company Sales data of lastb year')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(15,6))
month_number=df['month_number'].tolist()
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
toothpaste=df['toothpaste'].tolist()
bathingsoap=df['bathingsoap'].tolist()
shampoo=df['shampoo'].tolist()
moisturizer=df['moisturizer'].tolist()
plt.plot(month_number,facecream,label='facecream sales data',marker='o',linewidth=3)
plt.plot(month_number,facewash,label='facewash sales data',marker='o',linewidth=3)
plt.plot(month_number,toothpaste,label='toothpaste sales data',marker='o',linewidth=3)
plt.plot(month_number,bathingsoap,label='bathingsoap sales data',marker='o',linewidth=3)
plt.plot(month_number,shampoo,label='shampoo sales data',marker='o',linewidth=3)
plt.xlabel('Month Number')
plt.ylabel('Sales Unit in number')
plt.title('Sales Data')
plt.legend(loc='upper left')

```

```python
x=df['month_number'].tolist()
y=df['toothpaste'].tolist()
plt.scatter(x,y,label='toothpaste sales data',color='blue',)
plt.legend(loc='upper left')
plt.grid(color='black',linestyle='--',linewidth=0.5)
plt.xlabel('Month Number')
plt.ylabel('Sales Unit')
plt.title('Toothpaste sales data')
```

```python
plt.figure(figsize=(15,8))
month_number=df['month_number'].tolist()
facecream=df['facecream'].tolist()
facewash=df['facewash'].tolist()
plt.bar([a-0.25 for a in month_number],facecream,width=0.25,label='facecream Sales Data',align='edge')
plt.bar([a+0.25 for a in month_number],facewash,width=-0.25,label='facewash Sales Data',align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales unit in number')
plt.title('Facewash And Facecream Sales Data')
plt.grid(color='black',linestyle='--',linewidth=0.5)
plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=df['month_number'].tolist()
y=df['bathingsoap'].tolist()
plt.bar(x,y)
plt.xlabel('Month Number')
plt.ylabel('sales Unit ')
plt.title('bathing soap sales data')
plt.grid(color='black',linestyle='--',linewidth=0.5)
plt.show()
```

```python
plt.figure(figsize=(15,10))
x=df['total_profit']
y=[150000,175000,200000,225000,250000,300000,350000]
plt.hist(x,y,label='Profit data',color='red')
plt.xlabel('profit range in dollar')
plt.ylabel('actual profit in dollar')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(10,10))
month_number=df['month_number'].tolist()
mylabels=['Facecream','Facewash','Toothpaste','Bathing Soap','Shampoo','Moisturizer']
salesdata=[df['facecream'].sum(),df['facewash'].sum(),df['toothpaste'].sum(),df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]
plt.pie(salesdata,labels=mylabels,autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.show()
```

```python
plt.figure(figsize=(10,10))

month_number=df['month_number'].tolist()
bathingsoap=df['bathingsoap'].tolist()
facewash=df['facewash'].tolist()
plt.subplot(2,1,1)
plt.plot(month_number,bathingsoap,marker='o',color='black')
plt.title('Sales data of a Bathingsoap')
plt.subplot(2,1,2)
plt.plot(month_number,facewash,marker='o',color='red')
plt.xlabel('Month Number')
plt.ylabel('Sales unit in Number')
plt.title('Sales Data of Facewash')
plt.show()
```
